## Introduction ##

**Objective:** Find any trends between sleep score and active zone minutes per month in the year 2023.

**Background & Motivation:** 

FitBit data comes from tracking technology implemented in a wearable accessory. One aspect of this technology measures heart rate consistently whenever worn. Aside from visibly displaying this on the accessory's tiny monitor, FitBit also records and syncs heart rate (HR) into an accessible account. Within exercise science, there is a certain HR "zone" that is considered an ideal range for cardiovascular exercise.Similarly, FitBit has a proclaimed "Sleep Score" (where higher scores signifies better sleep). This sleep score is calculated by FitBit via variables such as HR, HR variability, oxygen saturation, breathing rate, etc.. 

**Questions to Explore:** 

What is the relationship between HR and Sleep? In particular, does sleep score correlate with exercise or does it not? What may be some reasons for the outcome?

## Methodology ##

**Overview:**

The project could be seperated into 3 components.

(1) Sleep Data - After ensuring the relevant files are in the same directory, I accessed csv files containing data on various sleep metrics representing a sleep score. Among them, sleep duration was selected. I then extracted this column using the pandas module, cleaned any missing or N/A values, and displayed the 2023 data in a simple line graph using matplotlib.

(2) Active Zone Data - This was the more complicated piece because the data for each month was scattered across 12 datasets (one for each month of 2023). I used various loop, indexing, and arithmatic techniques to quickly extract the relevant data and convert it into a dictionary of hours per month where the key was the month (expressed as an integer) and the respective values were the number of hours per month.

(3) Plot and Analysis - Sleep and active zone data were layered over one-another on a line graph (can be seen below) over the same date range (adjusted through indexing). From this, observations were made to determine what statistical techniques may be necessary to calculate a proxy of correlation. Also, conjecture was made as to what confounding variables may be impacting the data of either sleep or HR zone hours.

Note: This work will be provided in a seperate jupyter notebook.

**Independent/Dependent Variables:**

For an independent variable, the month range (correspending to the numerical month of the year) was calculated. The specific range: 7-12, was chosen due to the overlapping date requirements in the selected datasets. The dependent variables were the number of hours on each dimension (sleep duration & zone minutes). Because they the same units, I can layer them together in a graph.

**Statistics:** 

Two statistical measures were calculated from the two dependent variables: A correlation coefficient and a t-test statistic (producing a p-value). A correlation coefficient of 0.543 indicates a moderate positive linear relationship between the two datasets. This means that as one dataset increases, the other tends to increase as well, but the relationship is not extremely strong. Therefore, there is a postive correlation between sleep durtions and active zone minutes. Looking at the graph, it's clear that there is far less variability between the dependent variables. A simple t-test showed that the null hypothesis was not disproven with a p-value greater than .05

**Potential Confounds:** 

First, sleep requirements could effect the barrier to entering a HR zone. For instance, lack of sleep could increase HR throughout the day when performing lower intensity activity such as walking. The opposite may be true for periods of high rest. Second, the context of life circumstances (or simply differing desires/interests) may be influencing the fluctuations of the HR zone data. This could involve variations in the opportunity to perform dedicated HR zone exercise. Third, it is unclear how chronic emotional states may influence the dependent variables. For instance, in periods of stress of reletively high fight or flight signals, the body may respond with a higher heart rate in general.
