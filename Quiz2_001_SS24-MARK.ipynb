# 1. Writing a function (6 points)

In statistics, **arithmatic mean** or just **mean** or **average** is an important factor in calculating the Pearson correlation coefficient (PCC) which is a correlation coefficient that measures linear correlation between two sets of data. The PCC has a value between −1 and 1.

Mathematically,the equation for mean is
### $$\overline{x} =\frac{\sum_{i=1}^N x_i}{N} = \frac{x_1 + x_2 +\cdots + x_N}{N} $$


✅ **Write a function (6 points)**

- Write a function called `average` (**2 points**).
- Your function should take one input called: `x_values` (**2 point**).
- The function should return the mean value of x_values (**2 points**). 

```python
def average (x_values):
    return sum(x_values) / len(x_values)

average([1,2,3,4,5])
```

# 2. Visualizing a model using matplotlib (6 points)

Compound interest is interest accumulated from a principal sum and previously accumulated interest. It is the result of reinvesting or retaining interest that would otherwise be paid out, or of the accumulation of debts from a borrower. The total accumulated value, including the principal sum **P** plus compounded interest **I**, is given by the formula:

$$A= P(1+\frac{r}{n})^{nt}$$
where:

$A$ is the final amount

$P$ is the original principal sum

$r$ is the nominal annual interest rate

$n$ is the compounding frequency

$t$ is the overall length of time the interest is applied (expressed using the same time units as r, usually years).

Suppose an amount of $2,500.00 is deposited in a bank paying an annual interest rate of 3.8%, compounded monthly.

**Use the provided function, model parameters, and array `years`, to calculate a new array of the balance of each year**.

Then, **use matplotlib to make a plot of those balance values as a function of the `years` array.**

To check if the balance values seem plausible, **use the `plt.axvline()` function to make a line at year 6**, which should represent the balance after 6 years. 

**You should also make sure to label your x and y axes.**


```python
# Will need to use NumPy for the function
import numpy as np
# Will need to import matplotlib for the plotting
import matplotlib.pyplot as plt

# This function calculates the total accumulated value over a range of years
def total_accumulated_value(t, P, r, n):
     return P*(1 + r/n)**(n*t)

# Create year array
years = np.arange(0,30,1) + 1
```

```python
#np.array(total_accumulated_value(years, 2500, .038, 12))

data = total_accumulated_value(years, 2500, .038, 12)

plt.plot(data, label = "Compound Interest");
plt.xlabel("Year")
plt.ylabel("Monetary Value")
plt.title("Interest Over Time")
plt.axvline(x = 6, color = "r", label = "6 Year Balance")
plt.legend()
plt.show()
```
